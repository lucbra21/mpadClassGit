```python
from git import Repo

def check_git_status():
    repo = Repo('.')
    
    # Ver archivos modificados, nuevos y eliminados
    print("Cambios pendientes:")
    print("\nArchivos modificados:")
    for item in repo.index.diff(None):
        print(f"- {item.a_path}")
    
    print("\nArchivos sin seguimiento (untracked):")
    for item in repo.untracked_files:
        print(f"- {item}")

check_git_status()
```

```python
from git import Repo
from datetime import datetime
import os
from dotenv import load_dotenv

load_dotenv()

def hay_cambios(repo):
    return bool(repo.index.diff(None) or repo.untracked_files)

def git_push_changes(username, token):
    try:
        # Obtener el repositorio
        repo = Repo('.')
        
        # Verificar si hay cambios
        if not hay_cambios(repo):
            print("No hay cambios para subir")
            return
            
        # Si hay cambios, proceder con el push
        fecha_actual = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
        mensaje_commit = f'actualizacion automatica {fecha_actual}'
        
        # Add y commit
        repo.git.add('.')
        repo.git.commit('-m', mensaje_commit)
        
        # Construir la URL correctamente
        remote_url = f'https://{token}@github.com/{username}/mpadClassGit.git'
        
        # Push con la nueva URL
        repo.git.push(remote_url)
        
        print("Cambios subidos exitosamente a git")
        
    except Exception as e:
        print(f"Error al hacer push a git: {e}")
```
