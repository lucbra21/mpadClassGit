# GIT V2



Repaso de conceptos básicos:

1. Sistema de Control de Versiones:
- Git es un sistema que registra cambios en archivos a lo largo del tiempo
- Permite trabajo colaborativo y mantener diferentes versiones del código

2. Áreas principales:
- Working Directory: donde trabajas los archivos
- Staging Area (git add): área de preparación
- Local Repository (git commit): guarda los cambios confirmados
- Remote Repository (GitHub): almacenamiento en la nube

3. Comandos básicos revisados:
- `git add .`: prepara todos los cambios
- `git commit -m "mensaje"`: guarda cambios localmente
- `git push`: envía cambios a GitHub
- `git pull`: obtiene cambios desde GitHub

Conceptos nuevos a explicar:

1. Configuración Global:
```bash
git config --global user.name "nombre"
git config --global user.email "email"
```

2. Ramas (Branches):
- Son líneas independientes de desarrollo
- `git branch`: lista ramas
- `git branch nombre`: crea rama
- `git checkout nombre`: cambia a otra rama

3. Rebase:
- Reorganiza commits para mantener historia lineal
- `git rebase master`: integra cambios de master en rama actual

4. Resolución de Conflictos:
- Ocurren cuando hay cambios incompatibles
- Git marca los conflictos en archivos
- Se resuelven manualmente editando archivos

# Ejemplo errores si no hacemos la configuracion inicial


Los errores más comunes sin configuración global son:

```bash
*** Please tell me who you are.
Run
  git config --global user.email "you@example.com"
  git config --global user.name "Your Name"
```

Este error aparece al intentar hacer el primer commit sin configurar identidad.

También puedes ver:
```bash
fatal: unable to auto-detect email address
```

La solución es configurar las variables globales:
```bash
git config --global user.name "tu_nombre"
git config --global user.email "tu_email@ejemplo.com"
```

Si usas un email diferente al asociado con el repositorio privado en GitHub:

No podrás hacer push al repositorio
GitHub mostrará error de autenticación
Los commits aparecerán sin vincular a tu cuenta de GitHub

La solución es usar el email asociado a tu cuenta de GitHub o ser agregado como colaborador con el nuevo email.

# Ejemplo uso ramas

Aquí tienes un ejemplo práctico paso a paso:

```bash
# 1. Iniciar proyecto y primera versión
git init / o tambien usar git clone https://github.com/
git add .
git commit -m "versión inicial"

# 2. Crear y cambiar a rama de desarrollo
git branch desarrollo
git checkout desarrollo
# O más corto: git checkout -b desarrollo

# 3. Hacer cambios en rama desarrollo
touch archivo-nuevo.txt
git add .
git commit -m "agregado nuevo archivo"

# 4. Volver a master y hacer otros cambios
git checkout master
echo "cambios" > README.md
git commit -am "actualizado README"

# 5. Fusionar cambios de desarrollo en master
git checkout master
git merge desarrollo

# 6. Si hay conflictos
# Editar archivos manualmente
git add .
git commit -m "resueltos conflictos merge"

# 7. Subir cambios a GitHub
git push origin master
git push origin desarrollo
```

Para eliminar ramas:

```bash
# Eliminar rama local
git branch -d nombre_rama    # eliminación segura
git branch -D nombre_rama    # forzar eliminación

# Eliminar rama remota
git push origin --delete nombre_rama
```

# Ejemplo uso de reset

```
# Realizar cambios
echo "nuevo contenido" > archivo.txt
git add .
git commit -m "commit que queremos deshacer"

# Deshacer último commit manteniendo cambios en staging
git reset --soft HEAD~1

# Los cambios vuelven al staging area
# Podemos verificar con
git status

# Si queremos, podemos hacer un nuevo commit
git commit -m "nuevo mensaje de commit"
```

Otros tipos de reset:
- `--hard`: Elimina cambios completamente
- `--mixed`: Cambios vuelven a working directory

# Ejemplo checkout

```bash
# Ejemplo práctico de checkout

# 1. Checkout con archivos
git add archivo1.txt
git commit -m "cambios archivo1"
# Ver versión anterior del archivo
git checkout HEAD~1 archivo1.txt
# Descartar cambios en archivo
git checkout -- archivo1.txt

# 2. Checkout con ramas
git checkout -b feature1
echo "cambios feature1" > feature1.txt
git add .
git commit -m "cambios en feature1"

git checkout -b feature2
echo "cambios feature2" > feature2.txt
git add .
git commit -m "cambios en feature2"

# Moverse entre ramas
git checkout master
git checkout feature1
git checkout feature2

# 3. Checkout a un commit específico
git log --oneline  # ver hash de commits
git checkout abc123  # usar hash del commit deseado
# Volver a la rama principal
git checkout master
```

# Resolucion de conflictos

### Situación que genera conflicto
```
git checkout master
echo "cambios master" > archivo.txt
git commit -am "cambios master"

git checkout feature
echo "cambios feature" > archivo.txt
git commit -am "cambios feature"
```

### Al intentar merge aparece conflicto
```
git checkout master
git merge feature
```

### El archivo mostrará:
```
<<<<<<< HEAD
cambios master
=======
cambios feature
>>>>>>> feature
```

### Pasos para resolver:
1. Abrir archivo y elegir qué cambios mantener
2. Eliminar marcadores de conflicto (<<<<<<, =====, >>>>>>)
3. Guardar archivo
4. git add archivo.txt
5. git commit -m "resuelto conflicto"
